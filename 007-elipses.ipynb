# Elipses

Sejam $a > 0$, $F_1$ e $F_2$ pontos com $d(F_1, F_2) < 2a$. Uma elipse é o conjunto dos pontos $P$ com
$$
d(P,F_1) + d(P, F2) = 2a
$$

### Fazendo um gráfico pela definição

```python
import matplotlib.pyplot as plt
import numpy as np

all_x = np.arange(-4, 4, 0.1)
all_y = np.arange(-4, 4, 0.1)

elipse_x = []
elipse_y = []

F1 = np.array([ 1, 0])
F2 = np.array([-1, 0])

a = 3

for x in all_x:
    for y in all_y:
        if abs(np.linalg.norm(np.array([x,y]) - F1) + np.linalg.norm(np.array([x,y]) - F2) - 2*a) < 0.1:
            elipse_x.append(x)
            elipse_y.append(y)

plt.scatter(elipse_x, elipse_y)
plt.scatter([F1[0], F2[0]], [F1[1], F2[1]])
plt.xlim([-4,4])
plt.ylim([-4,4])

plt.show()
```

Se $F_1 = (1,0)$ e $F_2=(-1,0)$ e  $a=2$, temos que a equação cartesiana da elipse se dá por

$$
\frac{x^2}{4} + \frac{y^2}{3} = 1.
$$

Fazendo $x = 2\text{cos}(t)$ e $y = \sqrt{3}\text{sen}(t)$, podemos fazer novamente o gráfico utilizando estas equações paramétricas.

```python
import matplotlib.pyplot as plt
import numpy as np

all_x = np.arange(-4, 4, 0.1)
all_y = np.arange(-4, 4, 0.1)

elipse_x = []
elipse_y = []

F1 = np.array([ 1, 0])
F2 = np.array([-1, 0])

a = 2

for x in all_x:
    for y in all_y:
        if abs(np.linalg.norm(np.array([x,y]) - F1) + np.linalg.norm(np.array([x,y]) - F2) - 2*a) < 0.1:
            elipse_x.append(x)
            elipse_y.append(y)
            
t = np.arange(0, 2 * np.pi + 0.1, 0.1)
x = 2 * np.cos(t)
y = 3 ** 0.5 * np.sin(t)

plt.plot(x, y, color='r')
plt.scatter(elipse_x, elipse_y)
plt.scatter([F1[0], F2[0]], [F1[1], F2[1]])
plt.xlim([-4,4])
plt.ylim([-4,4])

plt.show()
```

## Translação

Novamente, façamos o gráfico da elipse acima, utilizando apenas equações paramétricas.

```python
t = np.arange(0, 2 * np.pi + 0.1, 0.1)
x = 2 * np.cos(t)
y = 3 ** 0.5 * np.sin(t)

plt.plot(x, y, color='r')

plt.xlim([-4,4])
plt.ylim([-4,4])

plt.show()
```

Para transladar a elipse na direção $u = (-1, 2)$, podemos simplesmente trocar $x$ por $x - 1$ e $y$ por $y + 2$. Vejamos:

```python
t = np.arange(0, 2 * np.pi + 0.1, 0.1)
x = 2 * np.cos(t)
y = 3 ** 0.5 * np.sin(t)

plt.plot(x, y, color='r', label="Original")
plt.plot(x - 1, y + 2, color='g', label="Transladada")

plt.xlim([-4,4])
plt.ylim([-4,4])

plt.legend()

plt.show()
```

## Rotação

A matriz de rotação $R_\theta$ é dada por 
$$
R_\theta = \left(
\begin{array}{cc}
    \text{cos}(\theta) & -\text{sen}(\theta) \\
    \text{sin}(\theta) & \text{cos}(\theta) \\
\end{array}
\right).
$$
Dessa forma, se quisermos rotacionar um ponto $P=(x,y)$ num ângulo $\theta$ e no sentido anti-horário, em torno da origem, precisamos substituir $P$ por $R_\theta P$.

```python
t = np.arange(0, 2 * np.pi + 0.1, 0.1)
x = 2 * np.cos(t)
y = 3 ** 0.5 * np.sin(t)

theta = np.pi / 4

plt.plot(x, y, color='r', label="Original")
plt.plot(np.cos(theta) * x - np.sin(theta) * y, np.sin(theta) * x + np.cos(theta) * y, color='g', label="Rotacionada")

plt.xlim([-4,4])
plt.ylim([-4,4])

plt.legend()

plt.show()
```

### Aplicações

```python
plt.figure(figsize=(8,8))

t = np.arange(0, 2 * np.pi + 0.1, 0.1)
x = 2 * np.cos(t)
y = 4 * np.sin(t)

n = 12
theta = np.pi / n


for i in range(0, 2 * n, 2):
    plt.plot(np.cos(i * theta) * x - np.sin(i * theta) * y, np.sin(i * theta) * x + np.cos(i * theta) * y, color='r')
    plt.plot(np.cos((i + 1) * theta) * x - np.sin((i + 1) * theta) * y, np.sin((i + 1) * theta) * x + np.cos((i + 1) * theta) * y, color='b')

plt.axis('off')
plt.show()
```
